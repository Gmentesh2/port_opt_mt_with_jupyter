```python
# Generate random weights (sum=1)
```

```python
weights = numpy.random.random(len(symbols))
weights = weights / numpy.sum(weights)
print(weights)
print("\n")
print(sum(weights))
```

```python
# Calculate return for portfolio
```

```python
numpy.sum(weights * returns.mean()) * 252
```

```python
# Calculate multyasset portfolio risk(standard deviation)
```

```python
numpy.sqrt(numpy.dot(weights.T, numpy.dot(returns.cov() * 252, weights)))
```

```python
# Analyzing returns and risks of 10000 combinations
```

```python
portfolio_return = []
portfolio_volatility = []
portfolio_sharp_ratio = []
portfolio_weight = []

for x in range(10000):
    # generate random weights
    p_weights = numpy.random.random(len(symbols))
    p_weights = p_weights / numpy.sum(p_weights)
    # p_weights /= numpy.sum(p_weights)
    
    # adding weights using those random weights
    
    return_1 = numpy.sum(p_weights * returns.mean() * 252 )
    portfolio_return.append(return_1)
    
    # adding volatility or standard deviation to list
    volatility_1 = numpy.sqrt(numpy.dot(p_weights.T, numpy.dot(returns.cov() * 252, p_weights )))
    portfolio_volatility.append(volatility_1)
    
    # get sharp ratio 
    sharp_ratio_1 = return_1 / volatility_1
    portfolio_sharp_ratio.append(sharp_ratio_1)
    
    # store weights for each portfolio 
    portfolio_weight.append(p_weights)
    
# convert to numpy array
portfolio_return = numpy.array(portfolio_return)
portfolio_volatility = numpy.array(portfolio_volatility)
portfolio_sharp_ratio = numpy.array(portfolio_sharp_ratio)
portfolio_weight = numpy.array(portfolio_weight)

```

```python
len(portfolio_weight)
```

```python
# plot efficient frontier
```

```python
# create dataframe with returns and volatility
portfolio = pandas.DataFrame({"Return": portfolio_return, "Volatility": portfolio_volatility})
# portfolio

portfolio.plot(x = "Volatility" , y = "Return" , kind = "scatter", figsize = (13,8))
```

```python
# return  the index of the largest sharp ratio 
```

```python
sharp_ratio_index = numpy.argmax(portfolio_sharp_ratio)
```

```python
sharp_ratio_index
```

```python
# find the ideal portfolio weights at that index 
```

```python
x = 0 
while x < len(symbols):
    print("Stock: %s : %2.2f" % (symbols[x],(portfolio_weight[sharp_ratio_index][x] * 100)))
    x += 1 

# find returns of the portfolio 
print("\nReturn:", portfolio_return[sharp_ratio_index])

# find volatility of the portfolio 

print("\nVolatility:" , portfolio_volatility[sharp_ratio_index])

# find sharp ratio of the portfolio 

print("\nSharp ratio:", portfolio_sharp_ratio[sharp_ratio_index])

    
```
